```python
import pandas as pd

# データ読み込み
nikkei225_df = pd.read_csv("../data/external/nikkei225.csv")
edinet_df = pd.read_csv("../data/external/EdinetcodeDlInfo.csv", encoding="cp932", skiprows=1)

# コードを文字列化して桁合わせ
nikkei225_df["銘柄コード"] = nikkei225_df["銘柄コード"].astype(str).str.zfill(4)
edinet_df["証券コード"] = edinet_df["証券コード"].astype(str).str.zfill(5)

# EDINETの証券コードから下1桁を落とす
edinet_df["証券コード_修正"] = edinet_df["証券コード"].str[:-1]

# マージ
merged_df = pd.merge(nikkei225_df, edinet_df, left_on="銘柄コード", right_on="証券コード_修正", how="left")
# 保存する場合
merged_df.to_csv("../data/processed/merged_nikkei225.csv", index=False)
```
